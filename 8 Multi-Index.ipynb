## Create a Multi-Index with the `.set_index()` Method

```python
df = pd.read_csv("bigmac.csv")
df["Date"] = pd.to_datetime(df["Date"]) 
# an alternative method to convert the column to datetime
df.head()
```

```python
df.nunique()
# date has the least number of unique values
```

```python
df.set_index(["Date","Country"], inplace = True)
# df.set_index(["Country", "Date"])
# multi-level index (two levels)
# the order of the list is important
# best practice (least unique number is in the outer level)
df
```

```python
df.sort_index(inplace = True)
# sorts the first index (dates), then second index (country), and so on...
# similar to .sort_values()
df
```

```python
df.index
# 2 indexes = tuple of 2 values
# 3 indexes = tuple of 3 values
# tuple = stores multiple items in a single variable

df.index.names
# immutable list
# index name is based on the column name in the dataframe
# they are also given an index position too (index within an index)
```

```python
type(df.index)

df.index[10]
# tuple
# type(df.index[10]) # to check
```

## Extract Index Level Values with the `.get_level_values` Method

```python
df = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
df.sort_index(inplace = True)
df.head(3)
```

```python
df.index.get_level_values("Date") # can use name of column 
df.index.get_level_values(0) # or use the index position
```

```python
df.index.get_level_values("Country")
df.index.get_level_values(1)
```

## Change Index Level Name with the `.set_names` Method

```python
df = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date","Country"])
df.sort_index(inplace = True)
df.head(3)
```

```python
df.index.set_names(names = ["Day","Location"], inplace = True) # changes the name of the index
df.head(3)
```

```python
df.index.set_names("Date",level = 0) 
df.index.set_names("Date", level = "Day", inplace = True)
# level of the multi index 
df.head(3)
```

## The `.sort_index()` Method on a `Multi-Index` `Data Frame`

```python
df = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date","Country"])
df.head(3)
```

```python
df.sort_index()
df.sort_index(ascending = True) # default
df.sort_index(ascending = False)
```

```python
# customizing the order of arrangement
df.sort_index(ascending = [True,False])
df.sort_index(ascending = [False,True], inplace = True)
```

```python
df.head()
```

```python
df.sort_index(level = 0)
df.sort_index(level = "Date")
df.sort_index(level = 1)
df.sort_index(level = "Country")
df.sort_index(level = "Country", ascending = False)
```

## The `.transpose()` Method

```python
df = pd.read_csv('bigmac.csv',parse_dates = ['Date'], index_col = ['Date','Country'])
df.sort_index(inplace = True)
df.head(3)
```

```python
df = df.transpose()
df.head(3)
```

```python
df.loc[('Price in US Dollars',),('2010-01-01','Sri Lanka'):('2010-01-01','Ukraine')] # pulling two indexes 
```

## The `.swaplevel()` Method

```python
df = pd.read_csv('bigmac.csv', parse_dates = ['Date'], index_col = ['Date','Country'])
df.sort_index(inplace = True)
df.head(3)
```

```python
df.swaplevel() # switches the index positions
#df.swaplevel('Country', 'Date')
df.swaplevel(0,1)
#df.swaplevel(1,0)
```

```python
df = df.swaplevel(1,0)
```

```python
df.head(3)
```

## The `.unstack()` Method

```python
df = pd.read_csv('worldstats.csv',index_col = ['country','year'])
df.head(3)
```

```python
s = df.stack()
s.head(3)
```

```python
s.unstack().unstack().unstack()
```

```python
s.unstack(2).unstack(0)
# selecting which index position you want to unstack
s.unstack(-1).unstack(-1)
```

```python
s.unstack('country')
```

```python
df = pd.read_csv('worldstats.csv',index_col = ['country','year'])
s = df.stack()
s.head(3)
```

```python
s.unstack(level = [0,1])
s.unstack(level = ['country','year'])
```

```python
s = s.unstack('year', fill_value = 0)
s.head()
```

## The `.pivot()` Method

```python
df = pd.read_csv('salesmen.csv', parse_dates = ['Date'])
df['Salesman'] = df['Salesman'].astype('category')
len(df)
```

```python
s = df.pivot(index = 'Date', columns = 'Salesman', values = 'Revenue')
len(s)
```

```python
s.info()
```

## The `.pivot_table()` Method

* like the 'summary' in jmp

```python
df = pd.read_csv('foods.csv')
df
```

```python
df.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'mean')
df.pivot_table(values = 'Spend', index = 'Gender', aggfunc = 'sum')
df.pivot_table(values = 'Spend', index = 'Item', aggfunc = 'sum')
df.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = 'City', aggfunc = ['count','sum','std'])
```

```python
pd.pivot_table(data = df, values = 'Spend', index = ['Gender','Item'], columns = 'City', aggfunc = 'sum')
```
