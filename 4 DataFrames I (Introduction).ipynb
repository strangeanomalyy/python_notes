## Broadcasting Operations

- tell the values in the series to do the exact same thing but they receive it individually 

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# Add
nba["Age"].add(5)
nba["Age"] + 5

# Subtract
nba["Salary"].sub(5000000)
nba["Salary"] - 5000000

# Multiply
nba["Weight"].mul(0.453592)
nba["Weight in Kilograms"] = nba["Weight"] * 0.453592
```

```python
nba.head(3)
```

```python
nba["Salary"].div(1000000)
nba["Salary in Millions"] = nba["Salary"] / 1000000
```

```python
nba.head(3)
```

## The `.astype()` Method

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(6)
```

```python
nba.dtypes # bec of null values, some of them are uploaded as float instead of integers
nba.info() # reduces memory usage if you change the dtypes (object → category; float → integer)
```

```python
nba["Salary"] = nba["Salary"].astype("int") 
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba["Weight"] = nba["Weight"].astype("int")
nba["Position"] = nba["Position"].astype("category")
nba["Team"] = nba["Team"].astype("category")
# float → integer
# there's no inplace parameter for .astype(), so equating it to itself overwrites the data
nba.head()
```

```python
nba["Position"].nunique() # number of unique values; unique() produces an array of the unique values
```

```python
nba["Team"].nunique()
```

## Sort `DataFrame` with the `.sort_values()` Method

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values("Name", ascending = False)
nba.sort_values("Age", ascending =  False)
nba.sort_values("Salary", ascending = False, inplace = True)
# NaN is sorted at the end
```

```python
nba.head()
```

```python
nba.sort_values("Salary", ascending = False, na_position = "first") # na_position = null values
```

#### Multiple columns for `.sort_values()` Method

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Team","Name"], ascending = [True, False], inplace = True)
nba.head(3)
# sorts team then name 
# [first col, second col] also applies for the ascending parameter
# can add as many columns as you want
```

## Rank Values with the `.rank()` Method

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head()
```

```python
nba.sort_values("Salary", ascending = False)
```
